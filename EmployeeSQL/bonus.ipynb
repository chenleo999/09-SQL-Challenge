```python
# read source csv files
import pandas as pd
department = pd.read_csv('./data/departments.csv')
dept_emp = pd.read_csv('./data/dept_emp.csv')
dept_manager = pd.read_csv('./data/dept_manager.csv')
employee = pd.read_csv('./data/employees.csv')
salary = pd.read_csv('./data/salaries.csv')
title = pd.read_csv('./data/titles.csv')

department.head(), dept_emp.head(), dept_manager.head(), employee.head(), salary.head(), title.head()
```

```python
# List the employee number, last name, first name, sex, and salary of each employee.
df_merge = employee.merge(salary, how='left', left_on='emp_no', right_on='emp_no')
df_display = df_merge[['emp_no', 'last_name', 'first_name', 'sex', 'salary']]
df_display
```

```python
# List the first name, last name, and hire date for the employees who were hired in 1986.
employee['hire_date'] = pd.to_datetime(employee['hire_date'])
df_select = employee[employee['hire_date'].dt.year == 1986]
df_display = df_select[['first_name', 'last_name', 'hire_date']]
df_display
```

```python
# List the manager of each department along with their department number, department name, employee number, last name, and first name.
df_merge1 = department.merge(dept_manager, how='left', left_on='dept_no', right_on='dept_no')
df_merge2 = df_merge1.merge(employee, how='left', left_on='emp_no', right_on='emp_no')
df_display1 = df_merge2[['dept_no', 'dept_name', 'emp_no', 'last_name', 'first_name']]
df_display2 = df_display1.rename(columns={'emp_no': 'manager_emp_no',
                           'last_name': 'manager_last_name',
                           'first_name': 'manager_first_name'})
df_display2
```

```python
# List the department number for each employee along with that employeeâ€™s employee number, last name, first name, and department name.
df_merge1 = employee.merge(dept_emp, how='left', left_on='emp_no', right_on='emp_no')
df_merge2 = df_merge1.merge(department, how='left', left_on='dept_no', right_on='dept_no')
df_display = df_merge2[['emp_no', 'last_name', 'first_name', 'dept_name']]
df_display
```

```python
# List first name, last name, and sex of each employee whose first name is Hercules and whose last name begins with the letter B.
df_select = employee[(employee['last_name'].str.startswith('B')) & (employee['first_name']=='Hercules')]
df_display = df_select[['first_name', 'last_name', 'sex']]
df_display
```

```python
# List each employee in the Sales department, including their employee number, last name, and first name.
df_merge1 = employee.merge(dept_emp, how='left', left_on='emp_no', right_on='emp_no')
df_merge2 = df_merge1.merge(department, how='left', left_on='dept_no', right_on='dept_no')
df_select = df_merge2[df_merge2['dept_name']=='Sales']
df_display = df_select[['dept_name', 'emp_no', 'last_name', 'first_name']]
df_display
```

```python
# List each employee in the Sales and Development departments, including their employee number, last name, first name, and department name.
df_merge1 = employee.merge(dept_emp, how='left', left_on='emp_no', right_on='emp_no')
df_merge2 = df_merge1.merge(department, how='left', left_on='dept_no', right_on='dept_no')
df_select = df_merge2[(df_merge2['dept_name']=='Sales') | (df_merge2['dept_name']=='Developmenet')]
df_display = df_select[['dept_name', 'emp_no', 'last_name', 'first_name']]
df_display
```
